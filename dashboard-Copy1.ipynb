```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

app.layout = html.Div([
    html.Div([dcc.Dropdown(id='party-dropdown',
            options=[
            {'label': 'Lewica', 'value': '__Lewica'},
            {'label': 'Polska 2050', 'value': 'PL_2050'},
            {'label': 'PO/KO', 'value': 'Platforma_org'},
            {'label': 'PiS', 'value': 'pisorgpl'},
            {'label': 'PSL', 'value': 'nowePSL'},
            {'label': 'Konfederacja', 'value': 'KONFEDERACJA_'}
        ],
        value=['__Lewica', 'pisorgpl', 'Platforma_org', 'KONFEDERACJA_', 'nowePSL', 'PL_2050'],
        multi=True
            ), 
            dcc.Graph(id='graph-with-slider')], 
             style={'width': '85%', 'float': 'right', 'display': 'inline-block'}),
    
    html.Div([
    html.B(html.Label('Choose aggregation:')),
    dcc.RadioItems(id='time-radio',
                    options=[
                        {'label': 'Daily', 'value': 'day'},
                        {'label': 'Weekly', 'value': 'week'},
                        {'label': 'Monthly', 'value': 'month'}
                    ],
                    value='week'),
    html.Br(),
    html.B(html.Label('Choose category:')),
    dcc.RadioItems(id='category-radio',
                  options=[
        {'label': 'Economy', 'value': 'economy'},
        {'label': 'Health', 'value': 'health'},
        {'label': 'Justice', 'value': 'justice'},
        {'label': 'Social issues', 'value': 'social'}              
                  ],
                  value='economy')], 
        style={'width': '15%', 'display': 'inline-block'})
])


@app.callback(
    Output('graph-with-slider', 'figure'),
    Input('party-dropdown', 'value'),
    Input('time-radio', 'value'),
    Input('category-radio', 'value'))

def update_figure(party, time, category):
    cursor.execute("""SELECT name, date_trunc('{}', created_at) AS creation_date, ROUND(AVG(score_{}),2) 
    FROM test2 
    WHERE screen_name IN ({})
    GROUP BY name, creation_date
    ORDER BY name, creation_date ASC;""".format(time, category, str(party).replace('[','').replace(']','')))
    a = cursor.fetchall()
    
    cols = ['name', 'created_at', 'score_' + category]
    df2 = pd.DataFrame(columns=cols)
    for i in range(len(a)):
        df2 = df2.append(pd.Series(a[i], index=cols), ignore_index=True)

    #parties = list(df2['name'].value_counts().index[:6])
    #df2 = df2[df2['name'].isin(parties)]
    #df2['score_' + category] = df2['score_' + category].apply(lambda x: random.uniform(0,1))

    fig = px.line(df2, x='created_at', y='score_' + category, color='name', 
                  title='Share of tweets containing a reference to the chosen category', hover_name='name',
                 labels={'created_at': 'Date', 'score_' + category : 'Share of tweets'}, 
                  hover_data={'name': False})
    fig.update_layout(transition_duration=500)
    fig.update_traces(mode='markers+lines')
    return fig

if __name__ == '__main__':
    app.run_server(debug=False)
```
